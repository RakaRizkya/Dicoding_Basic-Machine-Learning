```
# Import library yang dibutuhkan
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```
iris = load_iris()
atribut = iris.data
label = iris.target
# Membagi dataset menjadi train set dan test set
X_train, X_test, y_train, y_test = train_test_split(
    atribut, label, test_size=0.2)
```

```
# Membuat decision tree classifier tanpa PCA
tree_model = DecisionTreeClassifier()
model = tree_model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```
# Membuat PCA dengan 4 komponen (sesuai atribut)
pca = PCA(n_components=4)
# Mengaplikasikan PCA pada dataset
pca_attributes = pca.fit_transform(X_train)
# Melihat variance dari setiap atribut
pca.explained_variance_ratio_
```

```
# Membuat PCA dengan 2 komponen (total variance 2 atribut sudah cukup tinggi)
pca = PCA(n_components = 2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

```
# Pengujian setelah menggunakan PCA
model2 = tree_model.fit(X_train_pca, y_train)
model2.score(X_test_pca, y_test)
```
