```
# Import library yang dibutuhkan
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn.datasets import load_iris
```

```
# Load dataset iris
iris = pd.read_csv('Iris.csv')
display(iris.head())
# Membuang kolom id karena kurang penting pembuatan model
iris.drop('Id', axis=1, inplace=True)
display(iris.head())
```

```
# Memisahkan antara atribut dan label
X = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm' ]]
y = iris['Species']
```

```
# Membuat model decision tree
tree_clf = DecisionTreeClassifier()
# Melakukan pelatihan model terhadap data
tree_clf.fit(X, y)
```

```
# Melakukan tes pada model dengan memasukkan data baru
tree_clf.predict([[5.2, 0.4, 0.4, 0.3]])
```

```
# Melihat hasil visualisasi dengan library Graphviz
from sklearn.tree import export_graphviz
export_graphviz(
    tree_clf,
    out_file = "iris_flower.dot",
    feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
    class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica' ],
    rounded = True,
    filled = True
)
```
