```
# Import library yang dibutuhkan 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```
# Memuat dataset yang diperlukan
df = pd.read_csv('Social_Network_Ads.csv')
display(df.head())
display(df.info())
```

```
# Membuang kolom yang tidak diperlukan
data = df.drop('User ID', axis=1)
# Mengubah data kategorik menjadi kolom 
data = pd.get_dummies(data)
data
```

```
# Memisahkan antara atribut dengan label
X = data[['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']]
y = data['Purchased']
```

```
# Membagi data dengan train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```
# Membuat model logistic regression dan melatih model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```
# Mengukur score model
model.score(X_test, y_test)
```
