```
# Import library
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```
# Memuat dataset diabetes
df = pd.read_csv('diabetes.csv')
display(df.head())
display(df.info())
```

```
# Memisahkan atribut dan label pada dataset
X = df[df.columns[:8]]
y = df['Outcome']
```

```
# standarisasi nilai2 dari dataset
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```
# membagi train set dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```
# Membuat model Support Vector Classifier
clf = SVC()
clf.fit(X_train, y_train)
```

```
# Melihat score dari model yang telah dibuat
clf.score(X_test, y_test)
```
